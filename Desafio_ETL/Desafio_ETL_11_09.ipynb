```python
import pandas as pd

df = pd.read_csv("/content/dados_nao_tratados.csv")

df['idade'] = pd.to_numeric(df["idade"], errors="coerce")

df['nota'] = pd.to_numeric(df["nota"], errors="coerce")

df.fillna(0, inplace=True)

df['ativo'] = df['ativo'].replace({
    'sim': True,
    'Sim': True,
    'não': False,
    'Nao': False,
    'nao': False,
    'Não': False,
    'TRUE': True,
    'FALSE': False,
    'e': False,
    'Yes': True
    })

df['data_inscricao'] = df['data_inscricao'].replace({
    0: '2023-09-11',
    '2022-08-XX': '2022-08-01',
    '2023/13/01': '2023-12-01',
    '15/09/2022': '2022-09-15',
    '2023-02-30': '2023-02-28'
})

df['idade'] = df['idade'].replace({
    0.0: 20.0
})

df['nota'] = df['nota'].replace({
    0.0: 10.0
})

df['data_inscricao'] = pd.to_datetime(df['data_inscricao'], format='%Y-%m-%d', errors='coerce')

df
```

```python
df.to_csv('dados_tratados')
```
