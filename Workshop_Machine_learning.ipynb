**1. Abordar o problema e Analisar**

Objetivo - prever a espécie da flor com base nas medidas


**2. Obter os dados**

```python
# Manipulação e Geração de Dados
import pandas as pd

# Bibliotecas de Machine Learning
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, confusion_matrix
```

```python
df = pd.read_csv("/content/data_iris(in).csv")
df
```

**3. Explorar os Dados**

```python
df.info()
```

**4. Tratamento de dados**

Os tratamentos eram simples e fiz na tabela mesmo

**5. Separar Base de Dados em Arrays**

```python
x_data = df.iloc[:, 0:4].values
y_data = df.iloc[:, 4].values

x_data, y_data
```

**6. Técnicas de Pré-processamento**

```python
scaler = StandardScaler()
```

```python
preprocessor = Pipeline(steps=[
                                ('Scaler', scaler)
                            ]
                        )
```

```python
x_data = preprocessor.fit_transform(x_data)
x_data
```

**7. Dividir Base de Dados entre Treino e Teste**

```python
x_train, x_test, y_train, y_test = train_test_split(x_data, y_data, test_size=0.2)

x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

**8. Definir vários modelos e aplicar Treinamento**

```python
modelLogistic = LogisticRegression()
modelLogistic.fit(x_train, y_train)
```

```python
modelGaussianNB = GaussianNB()
modelGaussianNB.fit(x_train, y_train)
```

```python
modelTree = DecisionTreeClassifier()
modelTree.fit(x_train, y_train)
```

**9. Validar o Modelo**

```python
def exibir_metricas (y_true, predict, class_name):

    matriz_confucao_df = pd.DataFrame(
        confusion_matrix(y_true, predict),
        index=class_name,
        columns=class_name
    )

    # Imprime a Acurácia:
    print(f"A acurácia geral do modelo em relação a todos os dados foi de aproximadamente {accuracy_score(y_true, predict) * 100:.2f}%.")
    for i, label in enumerate(class_name):
        print(f"O modelo identificou corretamente {matriz_confucao_df.values[i][i]} registros com o rótulo '{label}'.")
```

**Modelo Logistic**

```python
predictLogistic = modelLogistic.predict(x_test)

exibir_metricas(y_test, predictLogistic, ['Setosa', 'Versicolor', 'Virginica'])
```

**Modelo GaussianNB**

```python
predictGaussianNB = modelGaussianNB.predict(x_test)

exibir_metricas(y_test, predictGaussianNB, ['Setosa', 'Versicolor', 'Virginica'])
```

**Modelo Decision Tree Classifier**

```python
predictTree = modelTree.predict(x_test)

exibir_metricas(y_test, predictTree, ['Setosa', 'Versicolor', 'Virginica'])
```

**10. Salvar a Solução**

```python
import joblib

joblib.dump(modelLogistic, 'modelLogistic.pkl')
joblib.dump(modelGaussianNB, 'modelGaussianNB.pkl')
```
